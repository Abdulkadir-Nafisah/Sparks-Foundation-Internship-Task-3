```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.displot(data2.Profit, color="blue", label="Profit", kde= False);
plt.ylim([0,500])
plt.legend()
```

From this visualization we can see that a large number of people did not make any profit

```python
sns.relplot(x='Sales', y='Profit', hue= 'Category', data=data2 )
```

From this visualization we can say the Tecnology Category are most likely to make the highest profits and sales

```python
plt.figure(figsize=(10,15))
sns.relplot(x='Sub-Category', y='Profit', hue= 'Category', data=data2 )
plt.xticks(rotation=90)
```

Looking into the Subcategory, we see that the Copies Subcategory make the highest profit followed by the blinders Subcategory.

```python
plt.figure(figsize=(10,15))
sns.relplot(x='Discount', y='Profit', hue= 'Ship Mode', data=data2 )
plt.xticks(rotation=90)
```

From this visualization we can say that the profit is not very much correlated to discount.

```python

sns.barplot(x='Segment', y='Profit', hue='Ship Mode', data=data2);
plt.figure(figsize=(10,15))
```

From this visualization we can see that Home Office Segment makes the highest profit most especially the First Class and Second Class Ship Mode
